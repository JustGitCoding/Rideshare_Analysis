```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Plot the data using MATLAB
plt.plot(x_axis,y_axis)
```

```python
# Plot the data using ax.plt() -- object oriented interface method
# Creates empty plot:
fig, ax = plt.subplots()
# Actually adds the data:
ax.plot(x_axis,y_axis)

# # fig, ax = plt.subplots() <--- this line can also be re-written as:
# fig = plt.figure()
# ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)

# If we do it this way we won't be able to change figure level attributes (axis name, title, etc) ?
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show() 
# Only do plt.show() once per session -- shows everything in new window (if there are more than one)
```

```python
# Create plot and add a label for the legend.
plt.plot(x_axis,y_axis,label='Boston')

# Create labels for x & y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Set y limit between 0 and 45
plt.ylim(0,45)

# Add a title
plt.title('PyBer Fare by Month')

#Add a legend
plt.legend()

```

```python
# Create plot and add a label for the legend.
plt.plot(x_axis,y_axis,marker='*',color='blue',linewidth=2,label='Boston')

# Create labels for x & y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Set y limit between 0 and 45
plt.ylim(0,45)

# Add a title
plt.title('PyBer Fare by Month')

# Add a grid
plt.grid()

#Add a legend
plt.legend()
```

```python
# Create plot with object oriented approach
fix, ax = plt.subplots()

# Plot the line
ax.plot(x_axis,y_axis,label = 'Boston', color = 'green', linewidth = 2, marker = 'D')

# Set y limits
ax.set_ylim(0,45)

# Create labels for axes + Title at the top
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')

# Add legend + gridlines
ax.legend()
ax.grid()
```

```python
# Creating BAR chart
plt.bar(x_axis,y_axis)
```

```python
# Create BAR chart
plt.bar(x_axis,y_axis,color='green',label='Boston')

# Label x & y axes
plt.xlabel('Date')
plt.ylabel('Fares($)')

# Add title
plt.title('PyBer Fare by Month')

# Add legend
plt.legend()
```

```python
# HORIZONTAL BAR chart
plt.barh(x_axis,y_axis, color='magenta', label='Boston')

# INVERT y-axis
plt.gca().invert_yaxis()

# Label x & y axes
plt.ylabel('Date')
plt.xlabel('Fare($)')

# Add title
plt.title('PyBer Fare by Month')

# Add legend
plt.legend()
```

```python
# BAR chart - object oriented interface method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)

```

```python
# HORIZONTAL BAR chart - object oriented interface method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='teal', label='Chicago')

# INVERT y-axis
ax.invert_yaxis()

# Label x & y axes
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')

# Add title
ax.set_title('PyBer Fare by Month')

# Add legend
ax.legend()
```

```python
# SCATTER PLOT using MATLAB method
# y_axis_larger = []
# for data in y_axis:
#     y_axis_larger.append(data*3)
# plt.scatter(y_axis,x_axis, c='r', label='Chicago', s=y_axis_larger)
plt.scatter(y_axis,x_axis, c='r', label='Chicago', s=[i*3 for i in y_axis])


# INVERT y-axis
plt.gca().invert_yaxis()

# Label x & y axes
plt.ylabel('Date')
plt.xlabel('Fare($)')

# Add title
plt.title('PyBer Fare by Month')

# Add legend
plt.legend()
```

```python
# SCATTER PLOT using Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c='skyblue', alpha=0.8, label='Boston', 
           edgecolors='black', linewidths=2,
           s=[i*5 for i in y_axis])


# INVERT y-axis
ax.invert_yaxis()

# Label x & y axes
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')

# Set limit for x values
ax.set_xlim(0,50)

# Add title
ax.set_title('PyBer Fare by Month')

# Add legend
ax.legend(loc='lower right')
```

```python
# PIE Charts - MATLAB method
plt.subplots(figsize=(8,8))
explode_values=(0,0,0,0,0,0,0.2,0,0,0,0,0)
colors=["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(y_axis, labels=x_axis, explode=explode_values, colors=colors, autopct='%.1f%%')
plt.show()
```

```python
# PIE Charts - Object-Oriented Interface method
explode_values=(0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors=["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis, labels=x_axis, explode=explode_values, colors=colors, startangle=90, counterclock=0, shadow=1, autopct='%.1f%%')
plt.show()
```
